```python
import re
import pandas as pd
```

```python
f = open('WhatsApp Chat with Vanshika.txt','r',encoding ='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
print(type(data))
```

```python
# rege1o1.com idhr ap apna message daldo and vo apko uska regre pattern dedega
pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages = re.split(pattern,data)[1:]
len(messages)
```

```python
dates = re.findall(pattern,data)
dates
```

```python
# Create DataFrame
df = pd.DataFrame({'user_message': messages, 'message_date': dates})
```

```python
df['message_date'] = df['message_date'].str.replace(' - ', '', regex=False)
df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%y, %H:%M')
```

```python
# Rename column
df.rename(columns={'message_date': 'date'}, inplace=True)
df.head()
```

```python
df.shape
```

```python
import re

users = []
messages = []

for message in df['user_message']:
    entry = re.split(r'([\w\W]+?):\s', message)
    if len(entry) > 2:
        users.append(entry[1])      # user name
        messages.append(entry[2])   # actual message
    else:
        users.append('group_notification')
        messages.append(entry[0])   # system or media message

df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)

df.head()
```

```python
# Assuming your datetime column is named 'date' and is in datetime64 format

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['time'] = df['date'].dt.time  # This gives time as datetime.time object
df.head()
```

```python
df.drop(columns='date',inplace=True)
```

```python
df['hour'] = df['time'].apply(lambda x: x.hour)
df['minute'] = df['time'].apply(lambda x: x.minute)
df['second'] = df['time'].apply(lambda x: x.second)

# Optionally drop the original time column
df.drop(columns='time', inplace=True)

df.head()
```
